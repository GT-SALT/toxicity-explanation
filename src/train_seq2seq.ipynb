```python
from seq2seq import JoinModel

model = JoinModel(CLASSIFIERS, CLASSIFIER_MODEL_NAME, \
                  SEQ2SEQ_MODEL_NAME, SEQ2SEQ_MODEL_NAME)

post = 'I thought drugs were the only things black people could shoot up Boy was I wrong.'
stereotype = 'do drugs; kill people; commit shootings'

outputs = model.forward(post, stereotype)
print(outputs.loss)
print(outputs.logits.shape)
```
